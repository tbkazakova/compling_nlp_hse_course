```python
# remove tags
def remove_tags(text):
    text_with_headings = re.sub('<br>', '\n', text)
    no_tags = re.sub('<.*?>', '', text_with_headings)
    good_text = re.sub('\n+', '\n', no_tags)
    return good_text
```
